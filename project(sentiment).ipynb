```python

import pandas as pd
df= pd.read_csv(r'D:\files\chrome_reviews.csv') #importing data
```

```python
df.head()
```

```python
df = df.drop(['ID','Review URL','Thumbs Up','User Name','Developer Reply',
                 'Version', 'Review Date', 'App ID'], axis=1)  #droping necessary columns
```

```python
df.head()
```

```python
import re
```

```python
df['review']=df['Text'].apply(lambda x:re.sub('[^A-Za-z0-9]+', ' ', str(x)))  #removing punctuation sign and emojis
```

```python
df.head()
```

```python
df.tail()
```

```python
df['review']=df['review'].str.lower()       #converting capital letters to small letters
```

```python
df.head()
```

```python
import nltk.corpus                 
```

```python
from nltk.corpus import stopwords
```

```python
stop=stopwords.words('english')         #creating instance for removing unnecessary words
```

```python

df['streview'] = df['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))    
    
    
```

```python
df.head()
```

```python
df.tail()
```

```python
from textblob import TextBlob
```

```python
def getPolarity(review):
    return TextBlob(review).sentiment.polarity    #checking polarity
```

```python
def analysis(score):
    if score < 0:
        return 'Negative'
    elif score == 0:
        return 'Neutral'
    else:
        return 'Positive'
```

```python
my_data = pd.DataFrame(df[['Text','Star','streview']])
```

```python
my_data['Polarity'] = my_data['streview'].apply(getPolarity) 
my_data['Analysis'] = my_data['Polarity'].apply(analysis)
```

```python
my_data.head()
```

```python
my_data.tail()
```

```python
m_data=my_data.loc[my_data['Analysis'] == 'Positive']     #selecting positive reviews
```

```python
dt=m_data.loc[m_data['Star']<3]
```
